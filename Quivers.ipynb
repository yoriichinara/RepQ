```python
import numpy as np
from tkinter import messagebox 

class IndQ: 

    # Cálculo de TODOS los caminos posibles 
    def walk(Q): 
        v = max(max(Q))
        for i in range(v+1): 
            for j in range(v+1):
                for k in range(v+1):  
                    if (i,j) in Q and (j,k) in Q: 
                        if Q.count((i,j)) == 1: 
                            Q.append((i,k))
                        else: 
                            for l in range(Q.count((i,j))): 
                                Q.append((i,k))
        for i in range(1, v+1): 
            if (i, i) in Q: 
                return print("!! You're quiver have oriented cycles")
            else: 
                return Q

    # Representaciones simples 
    def S_i(Q, i):
        v = max(max(Q))
        if i <= v and i > 0: 
            simpol = [0] * v 
            simpol[i-1] = 1
            #print(f'S({i}) = {simpol}')
        else: 
            messagebox.showwarning("Alert", "Revisa tu entrada")
        return simpol

    # Representaciones Proyectivas 
    def P_i(Q, i): 
        v = max(max(Q))
        if i <= v and i > 0: 
            proyec = [0] * v
            proyec[i-1] = 1
            for j in range(1, v+1):
                if (i,j) in Q: 
                    proyec[j-1] = Q.count((i,j))
            #print(f'P({i}) = {proyec}')
        else: 
            messagebox.showwarning("Alert", "Revisa tu entrada")
        return proyec

    # Representaciones Inyectivas
    def I_i(Q, i): 
        v = max(max(Q))
        if i <= v and i > 0: 
            inyec = [0] * v
            inyec[i-1] = 1 
            for j in range(1, v+1):
                if (j,i) in Q: 
                    inyec[j-1] = Q.count((j,i)) 
            #print(f'I({i}) = {inyec}')
        else: 
            messagebox.showwarning("Alert", "Revisa tu entrada")
        return inyec
    
    def tao_inv(Q, repres): 
        mac = max(max(Q))
        C = []

        for i in range(mac): 
            C.append(IndQ.P_i(Q, i+1))

        C = np.transpose(np.array(C))

        #print(C)
        C = -C.dot(np.transpose(np.linalg.inv(C)))

        return C.dot(np.transpose(np.array([repres])))
```

```python
# Ejemplos de Uso 
Q = [(1,2), (3,2), (4,3), (4,5)]

# Cálculo de caminos 
IndQ.walk(Q)

# Simples
print(f"S(4): {IndQ.S_i(Q, 4)}")

# Proyectivo
print(f"P(4): {IndQ.P_i(Q, 4)}")

# Inyectivo
print(f"I(2): {IndQ.I_i(Q, 2)}")

# Trasladado inverso
print(f"T^(-1) P(4): {[int(a) for a in np.transpose(IndQ.tao_inv(Q, [0, 1, 1, 0, 0]))[0].tolist()]}") 
```
